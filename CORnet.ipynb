```python
import os
import glob
import numpy as np
from scipy import stats
from scipy.special import softmax
from matplotlib import pyplot as plt
import pandas as pd
```

```python
feats = np.load("data/output/CORnet-Z_decoder_avgpool_feats.npy")
```

```python
# Get the filenames matching each row.
# (as in https://github.com/dicarlolab/CORnet/blob/d0cc17d4b34ad44dedb01683b70eafd15515adad/run.py#L216)
files = sorted(glob.glob(os.path.join("data/input", '*.*')))
```

```python
# Calculate probabilities over feature vector?
pr = softmax(feats, axis=1)
```

```python
plt.plot(pr.T)
plt.show()
```

```python
plt.plot(stats.entropy(pr, axis=1))
plt.show()
```

```python
scores = pd.DataFrame({"file": files, "entropy": stats.entropy(pr, axis=1)})
scores.to_csv("data/entropy_scores.csv")
```
